```python
import itertools
import numpy as np

def normalize(signal):
    return (signal - np.min(signal)) / (np.max(signal) - np.min(signal))

def discretize(signal):
    return list(map(lambda x: round(x), signal))

def differentiate(signal):
    s1, s2 = itertools.tee(signal)
    next(s2)
    return (y-x for x, y in zip(s1, s2))

def remove_ripple(stream, width):
    stream = [round(val) for val in stream]
    for idx in range(len(stream)):
        yield stream[idx]
        if idx+width >= len(stream): continue
        if stream[idx] and stream[idx+width]:
            stream[idx:idx+width] = itertools.repeat(1.0, width)

def stream_to_spikes(stream):
    ds = differentiate(itertools.chain(iter([0.0]), stream, iter([0.0])))
    ds1, ds2 = itertools.tee(ds)
    begins = (idx for idx, val in enumerate(ds1) if val == 1)
    ends = (idx for idx, val in enumerate(ds2) if val == -1)
    return zip(begins, ends)

def spikes_to_points(spikes):
    for begin, end in spikes:
        yield int((begin + end) / 2)

def stream_to_points(stream, tolerance=3):
    sanitized_stream = remove_ripple(stream, tolerance)
    spikes = stream_to_spikes(sanitized_stream)
    return spikes_to_points(spikes)
```
